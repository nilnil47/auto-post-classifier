```python
def generate_grade(rank: pd.Series):
    weights = {
    "antisemitism": 0.15,
    "graphic_violence": 0.15,
    "weapons": 0.10,
    "call_for_violence_operation": 0.10,
    "political_content": 0.15,
    "supporting_in_terror": 0.15,
    "misinformation": 0.20
}
    grade = 0
    for dimension in weights:
        grade += rank[dimension + "_rnk"] * weights[dimension]
    
    rank['grade'] = grade
    return rank


results.apply(generate_grade, axis=1)
```
